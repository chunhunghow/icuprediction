### Curate patient admissions data

Get index ICU admission, but remove those short first stay (<4h)
1. Remove those with just single admission (no readmission) and its a short ICU stay (<4h), called dataset indx_pat
2. For the cases of multiple admission, if their first admission is a short stay, then merge it with second.

```python
#(1) Remove short stay with just single admission
indx_pat = pd.DataFrame()
for idx in pat['PAT_ENC_CSN_ID'].unique():
    temp = pat[pat['PAT_ENC_CSN_ID'] == idx].sort_values(by='ICU Admission Time').reset_index(drop=True)
    temp['adm_idx'] = pd.Series(np.arange(temp.shape[0]))
    indx_pat = pd.concat([indx_pat, temp])
indx_pat = indx_pat.reset_index(drop=True)
temp = indx_pat.groupby('PAT_ENC_CSN_ID').agg({'adm_idx':max,'ICU LOS':'max'}).reset_index()
indx_pat = indx_pat[~indx_pat['PAT_ENC_CSN_ID'].isin(temp[(temp['adm_idx'] == 0) & (temp['ICU LOS']<=4)]['PAT_ENC_CSN_ID'].values)]


print(f'Original {len(pat["PAT_ENC_CSN_ID"].unique())} Curated {len(indx_pat["PAT_ENC_CSN_ID"].unique())}')
```

A function to compute Hosp LOS after ICU

```python

def stay_after_icu(enc_id :int , pat : pd.DataFrame):
    """To compute the LOS after ICU stay, given discharge time and hospital discharge time.
    Some do not have hospital discharge time, compute hospital LOS (if exists) + ICU discharge,
    else take discharge time as hosp discharge.
    ASSUME ICU DISCHARGE TIME NOT NULL
    """
    ## for readmission case
    data = pat[pat['PAT_ENC_CSN_ID'] == enc_id].sort_values(by='ICU Admission Time')
    assert len(data) != 0
    
    inp_dtime = data['INP_ADM_DATE'].unique()[0] #must exists
    
    #hospital discharge time is same for multiple icu admission patient
    if data['HOSP_DISCH_TIME'].isnull().all():
        if data['Hospital LOS'].notnull().all():
            data['HOSP_DISCH_TIME'] =pd.to_datetime(data['ICU Discharge Time'].unique()[-1]) + dt.timedelta(data['Hospital LOS'].unique()[0]) #last discharge time
        else: #both H LOS and H Discharge missing
            data['HOSP_DISCH_TIME'] = pd.to_datetime(data['ICU Discharge Time'].unique()[-1])
        

    
    in_out_dtime = data[['ICU Admission Time', 'ICU Discharge Time']].values
#     hospital_los = data['Hospital LOS'].values 
    disc_dtime = data['HOSP_DISCH_TIME'].values[0]
    n_adms = len(in_out_dtime)
    hosp_after_icu = []

    if n_adms > 1:
        for i in range(n_adms-1):
            hosp_after_icu += [np.timedelta64(in_out_dtime[i+1][0] - in_out_dtime[i][1], 'h')] 

    hosp_after_icu += [ np.timedelta64(disc_dtime - in_out_dtime[-1][1],'h')] 
    return hosp_after_icu


#ANALYSIS

# temp = indx_pat.groupby('PAT_ENC_CSN_ID').agg({'adm_idx':max,'ICU LOS':'max'}).reset_index()
# temp = indx_pat[indx_pat['PAT_ENC_CSN_ID'].isin(temp[(temp['adm_idx'] > 0)]['PAT_ENC_CSN_ID'])] 
# analysis = indx_pat[indx_pat['PAT_ENC_CSN_ID'].isin(temp[(temp['adm_idx'] == 0) & (temp['ICU LOS'] <=4)]['PAT_ENC_CSN_ID'].values) & (indx_pat['adm_idx']<2)]



# gap_between_readmission = []
# for id in analysis['PAT_ENC_CSN_ID'].unique():
#     gap_between_readmission += [stay_after_icu(id,analysis)[0].item().seconds/ 60 /60]
    
```

Iterate every encounters, get their hospital stay after ICU using function above, save in variable 'store'

```python
### First, compute their LOS after discharge and before hospital discharge or readmission

temp_pat = indx_pat.copy()
store = pd.DataFrame()
for enc_id in temp_pat['PAT_ENC_CSN_ID'].unique():
    d = temp_pat[temp_pat['PAT_ENC_CSN_ID']==enc_id].sort_values('ICU Admission Time').reset_index(drop=True)
    d['hosp_after_icu'] = pd.Series(stay_after_icu(enc_id,pat))
    store = pd.concat([store, d])
del d
store['hosp_after_icu'] = store['hosp_after_icu'].apply(lambda x: x.days*24 + x.seconds/60/60)
```

2. By now there is no short single admission, only those short first admission with readmission, iterate all encs and merge those short first admission, save to pat_cohort. pat_cohort will be the curated patient admission cohort dataset.

3. If there are still < 4 total ICU duration (after merging multiple stay), remove again.

```python
##(2) To MERGE : go through all ID and edit their LOS and admission time

temp = indx_pat.groupby('PAT_ENC_CSN_ID').agg({'adm_idx':max,'ICU LOS':'max'}).reset_index()
temp = indx_pat[indx_pat['PAT_ENC_CSN_ID'].isin(temp[(temp['adm_idx'] > 0)]['PAT_ENC_CSN_ID'])] #pat data with multiple adm

indx_pat2 = pd.DataFrame()

for id in store['PAT_ENC_CSN_ID'].unique():
    if id in temp[(temp['adm_idx'] == 0) & (temp['ICU LOS'] <=4)]['PAT_ENC_CSN_ID'].values: #take those first adm less than 4 hours
        tobemerged = store[store['PAT_ENC_CSN_ID'] == id].sort_values(by='adm_idx').reset_index(drop=True)
        first_adm = tobemerged.iloc[0,:]
        other_adm = tobemerged.iloc[1:,:].reset_index(drop=True)
        other_adm['ICU Admission Time'].at[0] = first_adm['ICU Admission Time']
        other_adm['ICU LOS'].at[0] = other_adm['ICU LOS'].at[0] + first_adm['ICU LOS']
        indx_pat2 = pd.concat([indx_pat2, other_adm])
    else:
        indx_pat2 = pd.concat([indx_pat2, store[store['PAT_ENC_CSN_ID'] == id]])
        
indx_pat2 = indx_pat2.reset_index(drop=True)
pat_cohort = indx_pat2.copy()

pat_cohort = pd.merge(pat_cohort, pat_cohort.groupby('PAT_ENC_CSN_ID').agg({'adm_idx':'min'}).reset_index().rename(columns={'adm_idx':'min_adm_idx'}),on='PAT_ENC_CSN_ID')

#2
#remove short stay again after merging admission
pat_cohort = pat_cohort.drop(pat_cohort[(pat_cohort['adm_idx'] == pat_cohort['min_adm_idx']) & (pat_cohort['ICU LOS'] < 4)].index)
```

```python
import scipy
from scipy.stats import ttest_ind
from sklearn.metrics import confusion_matrix

statsfordan = pd.read_csv("C:/Users/chunhunghow/Project Repository/ICUMV in D/repo/ckpt/statistics/cohort.csv")
# statsfordan = statsfordan[statsfordan['mv_duration']>4]

def composite(rows):
    if (rows['hosp_after_icu'] > 168*2) | (rows['mortality']==1) | (rows['readmission']==1):
        return 1
    else: return 0
    


statsfordan['gender male'] = statsfordan['gender'] == 'Male'
# statsfordan['gender female'] = statsfordan['gender'] == 'Female'
statsfordan['R BMI'] = statsfordan.apply(lambda row: row['WEIGHT/SCALE']/16 / row['HEIGHT']**2 * 703,1 )
statsfordan['vasopresser'] = statsfordan['vasopresser'] > 0
statsfordan['discharge'] = statsfordan['mortality']
statsfordan['mortality'] = statsfordan['mortality'].apply(lambda x: 'Death' in x)

# resp_neg_col = [col for col in statsfordan.columns if ('resp_secretion_24' in col) & ('Small' not in col) & ('Large' not in col)]
statsfordan['resp_secretion_24_pos'] = statsfordan['resp_secretion_24_Large'].fillna(0) + statsfordan['resp_secretion_24_Moderate'].fillna(0)
# statsfordan['resp_secretion_neg'] = statsfordan[resp_neg_col].fillna(0).sum(1)
statsfordan['resp_secretion_24_pos'] = statsfordan['resp_secretion_24_pos'] > 0
# statsfordan['resp_secretion_neg'] = statsfordan['resp_secretion_neg'] > 0



statsfordan['resp_secretion_6_pos'] = statsfordan['resp_secretion_6_Large'].fillna(0) + statsfordan['resp_secretion_6_Moderate'].fillna(0)
statsfordan['resp_secretion_6_pos'] = statsfordan['resp_secretion_6_pos'] > 0




statsfordan['composite_risk'] = statsfordan.apply(lambda row: composite(row),1).values


TOTAL_TREATMENT =  statsfordan['composite_risk'].sum()
TOTAL_CONTROL = (statsfordan['composite_risk'] == 0).sum()



# def pval(col_control, col_treat , test):
#     if test == 'median':
        
#         med_control = col_control.median()
#         lower = col_control.quantile(0.25)
#         upper = col_control.quantile(0.75)
#         iqr_control = upper - lower
        
#         med_treat = col_treat.median()
#         lower = col_treat.quantile(0.25)
#         upper = col_treat.quantile(0.75)
#         iqr_treat = upper - lower
        
        
#         a = col_control.values
#         b = col_treat.values
#         whitney = scipy.stats.mannwhitneyu( a,b, alternative='two-sided') # method='exact'
#         return (med_treat, iqr_treat,med_control,iqr_control  , whitney.pvalue)
    
    
#     elif test == 'n%':
        
#         n_control = len(col_control)
#         n_treat = len(col_treat)

#         perc_treat = n_treat / TOTAL_TREATMENT
#         perc_control = n_control / TOTAL_CONTROL
        
#         pearson = scipy.stats.chi2_contingency([n_control, n_treat]) #if f_exp is None, then mean of f_obs is taken , equally likely
#         return (n_treat, perc_treat, n_control, perc_control, pearson.pvalue)
    
#     else:
#         raise NotImplemented

        
        
def chi2test(varA, varB):
    
    # lets fix varB as risk column
    
    cm = confusion_matrix(varA,varB)
    pval=scipy.stats.chi2_contingency(cm)[1]
    n_control, n_treat = cm[1,:]
    perc_treat = n_treat / TOTAL_TREATMENT
    perc_control = n_control / TOTAL_CONTROL
    return (n_treat, perc_treat, n_control, perc_control, pval)

def mediantest(col_control, col_treat):
    med_control = col_control.median()
    lower = col_control.quantile(0.25)
    upper = col_control.quantile(0.75)
    iqr_control = upper - lower

    med_treat = col_treat.median()
    lower = col_treat.quantile(0.25)
    upper = col_treat.quantile(0.75)
    iqr_treat = upper - lower


    a = col_control.values
    b = col_treat.values
    whitney = scipy.stats.mannwhitneyu( a,b, alternative='two-sided') # method='exact'
    return (med_treat, iqr_treat,med_control,iqr_control  , whitney.pvalue)




var_test = {'age': 'median',
 'gender male' : 'n%',
#  'gender female' : 'n%',
 'chronic' : 'n%',
 'cci' : 'median',
 'R BMI' : 'median' ,
 'adm_apache' : 'median',
 'disc_sofa' : 'median',
 'disc_apache' : 'median',
 'adm_sofa' : 'median',
 'vasopresser': 'n%',
 'comorb_Heart Failure' : 'n%',
 'comorb_Stroke' : 'n%',
 'comorb_Others: TB/ ILD/ Chest wall deformity/OSA/ Bronchiectasis': 'n%',
 'comorb_Pneumonia' : 'n%',
 'comorb_Kidney' : 'n%',
 'comorb_Cirrhosis' : 'n%',
 'comorb_COPD' : 'n%',
 'comorb_Immuno' : 'n%',
 'mv_duration' : 'median',
 'icu los':'median',
 
 ##discharge
 'PULSE' : 'median',
 'RESPIRATIONS' : 'median',
 'R CPN GLASGOW COMA SCALE SCORE' : 'median',
 'R CPN GLASGOW COMA SCALE BEST MOTOR RESPONSE' : 'median',
 'resp_secretion_24_pos' : 'n%',
 'resp_secretion_6_pos' : 'n%',
#  'resp_secretion_pos' : 'n%',
 'input/output': 'median',
 'FIO2, ABG' : 'median',
 'PULSE OXIMETRY' : 'median', #SPO2
 'POCT PH' : 'median',
 'POCT PO2' : 'median',
 'PO2/FIO2' : 'median',
 'POCT PCO2' : 'median',
 'BICARBONATE, SERUM' : 'median',
}



stats_result = []
for k,v in var_test.items():
    if statsfordan[k].dtype in (float, np.int64) :
        tmp = statsfordan[~statsfordan[k].isna()]
    if v == 'n%':
        stats_result += [[k,*chi2test(tmp[k].fillna(False), tmp['composite_risk'])]]
    elif v == 'median':        
        stats_result += [[k,*mediantest(tmp[(tmp['composite_risk']==0) ][k], tmp[(tmp['composite_risk']==1)][k] )]]
    


```
